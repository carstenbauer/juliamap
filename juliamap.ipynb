```julia
function getlatlong(s::AbstractString)
    location = geolocator[:geocode](s)
    location==nothing ? nothing : (location[:latitude], location[:longitude])
end
latlongs = getlatlong.(places[:name])
```

```julia
nothings = Dict{Int, String}()
for i in find(x->x==nothing, latlongs)
    nothings[i] = places[:name][i]
end
nothings
```

```julia
manual = Dict(
    2 => (-9.555682, -35.777877),
    21 => (41.907984, 12.495731),
    7 => (51.222810, 4.410232),
    17 => (50.928162, 6.928819),
    6 => (34.068921, -118.445181),
    24 => (41.903763, 12.514438),
    4 => (-22.768476, -43.685035),
    3 => (52.208872, 21.008658),
    5 => (-22.979107, -43.233083),
    23 => (-18.918506, -48.258170),
    12 => (22.520360, 75.920723)
)

for (k,v) in manual
    latlongs[k] = v
end
coords = hcat(getindex.(latlongs, 1), getindex.(latlongs, 2))
places[:lat] = coords[:,1]
places[:lon] = coords[:,2]
places
```

```julia
# Make an empty map
m = folium.Map(location=(20, 0), tiles="Mapbox Bright", zoom_start=2)
 
# Add city markers
for i in 1:nrow(places)
    folium.Marker([places[i,:lat], places[i,:lon]], popup=string(places[i,:name]))[:add_to](m)
end
 
# Save it as html
m[:save]("juliamap.html")
m
```
